```python
import random
import copy
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
# from dwave.system import DWaveCliqueSampler
from openjij import SQASampler
```

```python
QUBO = np.zeros((8, 8))
index_slices = [(i, i+4) for i in range(0, 8, 4)]
print(index_slices)
for b, e in index_slices:
    for m in range(b, e):
        for n in range(b, e):
            if m < n:
                QUBO[m, n] -= 1
            if m == n:
                QUBO[m, n] += 2*1

fig, ax = plt.subplots()
im = ax.imshow(QUBO, cmap="Blues")
fig.colorbar(im, ax=ax)
plt.show()
```

```python
import neal
sampler = neal.SimulatedAnnealingSampler()
sampleset = sampler.sample_qubo(QUBO,num_reads=1000)
```
